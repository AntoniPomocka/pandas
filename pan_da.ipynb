```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
prices = [
    (1, 2.12),
    (2, 2.56),
    (3, 3.10),
    (4, 3.16),
    (5, 3.58),
    (6, 5.12),
    (7, 5.16),
    (8, 5.20),
    (9, 4.12),
    (10, 4.10),
    (11, 3.65),
    (12, 4.25),
]
```

```python
df = pd.DataFrame(prices, columns=["Month", "PricePLN"])
df.set_index("Month", inplace=True)
df["PriceUSD"] = df["PricePLN"] / 4

print(df)

plt.plot(df.index, df["PriceUSD"], 'r--', label="Price in USD")
plt.xlabel("Month")
plt.ylabel("Price")
plt.title("Product Prices in USD")



plt.yticks(np.arange(0, 2.25, 0.25))

plt.show()
```
